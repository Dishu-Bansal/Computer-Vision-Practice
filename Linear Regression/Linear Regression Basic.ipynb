```python
import torch
import torch.nn as nn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("D:/AI/Pytorch Practice/Linear Regression/dataset/train.csv")
test = pd.read_csv("D:/AI/Pytorch Practice/Linear Regression/dataset/test.csv")
dataset.head()
X = dataset['x'].to_numpy(np.int)
ind = np.where(X == 3530)
print(ind)
X = np.delete(X, ind)
Y = dataset['y'].to_numpy()
Y = np.nan_to_num(Y)
Y = np.delete(Y, ind)
X_test = test['x'].to_numpy()
Y_test = test['y'].to_numpy()
print(X, Y)
```

```python
N = len(X)
N_test = len(X_test)
N
```

```python
plt.scatter(X, Y)
```

```python
X = X.reshape(N, 1)
Y = Y.reshape(N, 1)
X_test = X_test.reshape(N_test, 1)
Y_test = Y_test.reshape(N_test, 1)

inputs = torch.from_numpy(X.astype(np.float32))
targets = torch.from_numpy(Y.astype(np.float32))
```

```python
epochs = 500
losses = []
learning_rate = 10e-4

model = nn.Linear(1, 1)
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
for it in range(epochs):
    optimizer.zero_grad()
    
    outputs = model(inputs)
    loss = criterion(outputs, targets)
    
    losses.append(loss.item())
    
    loss.backward()
    optimizer.step()
    
    print(f'Epoch {it+1}/{epochs}, Loss: {loss.item():.4f}')
    
```

```python
plt.plot(losses)
```

```python
predicted = model(inputs).detach().numpy()
plt.scatter(X, Y, label='Original data')
plt.plot(X, predicted, label='Fitted line', color='red')
plt.legend()
plt.show()
```
